# Hipotesis:

Hipótesis nula (H₀):
No existe una relación estadísticamente significativa entre las variables masa corporal, tensión arterial, peso y medida de cintura y el riesgo de hipertensión en la población analizada.

Hipótesis alternativa (H₁):
Existe al menos una relación estadísticamente significativa entre alguna de las variables masa corporal, tensión arterial, peso o medida de cintura y el riesgo de hipertensión en la población estudiada.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[[ 'peso', 'medida_cintura', 'tension_arterial', 'masa_corporal']]
Y = df['riesgo_hipertension']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42 )

x_train_constante = sm.add_constant(X_train)
x_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, x_train_constante).fit()

Yc = modelo.predict(x_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinación: {r2: 0.4}')

from statsmodels.formula.api import ols

moedo_2 = ols(formula = 'riesgo_hipertension ~ peso + medida_cintura + tension_arterial + masa_corporal', data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)  
tabla_anova
```

```python
b0, b1, b2, b3, b4,  = modelo.params

Fun = lambda X1, X2, X3, X4: b0 + b1 * X1 + b2 * X2 + b3 * X3 + b4 * X4 

#Eres el encargado de evaluar a un paciente con sospecha de hipertensión. Durante un chequeo médico inicial, registras la siguiente información relevante.
#Con base en estos datos, ¿es probable que el paciente presente hipertensión?

Fun(90, 115, 97, 34.22)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(modelo.resid, dist="norm", plot=plt)
plt.title( 'Redisuales')
plt.show()

plt.hist(modelo.resid, bins=30)
plt.title("Histograma de los residuos")
plt.xlabel("Residuo")
plt.ylabel("Frecuencia")
plt.show()
```

```python
from scipy.stats import shapiro

stat, p = shapiro(modelo.resid)
print('Estadístico Shapiro-Wilk=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print("Los residuos parecen normales.")
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan

_, pval, __, ___ = het_breuschpagan(modelo.resid, modelo.model.exog)
print("p-valor de Breusch-Pagan:", pval)
```

# Evaluación clínica preliminar
Como parte del proceso de evaluación clínica, se recibe a un paciente con sospechas de padecer hipertensión. Durante el chequeo médico inicial, se registran datos relevantes como peso, medida de cintura, tensión arterial y masa corporal. Con base en esta información, se procede a analizar la probabilidad de que el paciente sea hipertenso utilizando un modelo de regresión previamente desarrollado.

# Interpretación general de los resultados
A partir del análisis de varianza (ANOVA), se evaluó el efecto de cuatro variables predictoras —peso, medida de cintura, tensión arterial y masa corporal— sobre una variable dependiente relacionada con el riesgo de hipertensión. El coeficiente de determinación obtenido (R² = 0.3059) indica que el modelo explica aproximadamente el 30.59% de la variabilidad en la variable de respuesta, lo que representa un nivel moderado de ajuste.

Los resultados revelan que tres de las cuatro variables presentan una relación estadísticamente significativa con la variable dependiente, al presentar valores-p menores al umbral de significancia convencional (α = 0.05):

Medida de cintura (p = 1.27e-05)

Tensión arterial (p = 2.09e-60)

Masa corporal (p = 4.90e-24)

En contraste, la variable peso (p = 0.3146) no mostró un efecto significativo dentro del modelo.

Dado que al menos una variable independiente resulta estadísticamente significativa, se rechaza la hipótesis nula (H₀). Por lo tanto, se concluye que existe evidencia suficiente para afirmar que al menos una de las variables estudiadas influye de forma significativa en el riesgo de hipertensión en la muestra analizada.

# Interpretación del resultado aplicado al paciente
Al ingresar los datos clínicos del paciente en el modelo de regresión, se obtiene como resultado una alta probabilidad de hipertensión, lo cual confirma las sospechas iniciales. El modelo sugiere que el paciente podría ya presentar hipertensión o encontrarse en un estado de riesgo elevado.

# Análisis de supuestos del modelo
Normalidad de los residuos
La prueba de Shapiro-Wilk fue utilizada para verificar el supuesto de normalidad en los residuos del modelo. El resultado obtenido (p < 0.05) indica que los residuos no se distribuyen normalmente, por lo que se rechaza la hipótesis nula. Esto constituye una violación del supuesto de normalidad, lo cual podría afectar la validez de las inferencias.

Homoscedasticidad
También se evaluó la homocedasticidad, es decir, la constancia de la varianza de los errores. Dado que el p-valor fue prácticamente cero, se concluye que los residuos presentan heterocedasticidad. Esta condición implica que la varianza de los errores no es constante, lo que representa otra violación a los supuestos del modelo de regresión lineal clásica.

# Conclusión general
El análisis realizado permitió identificar variables significativamente asociadas al riesgo de hipertensión, destacando la importancia de la medida de cintura, la tensión arterial y la masa corporal como predictores relevantes. Aunque el modelo logró explicar una parte considerable de la variabilidad en la variable dependiente (R² = 0.3059), también se detectaron violaciones a los supuestos de normalidad y homoscedasticidad, lo que sugiere la necesidad de considerar alternativas metodológicas más robustas, como modelos no paramétricos o transformaciones de variables.

Finalmente, al aplicar el modelo a un caso clínico, se identificó una alta probabilidad de hipertensión, lo cual confirma su utilidad como herramienta de apoyo en la detección temprana de riesgos de salud, siempre y cuando se interprete con cautela y se complementen los resultados con otras evaluaciones clínicas.
